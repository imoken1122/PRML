```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg

def f(x):
    return np.sin(2*np.pi*x)

def dataset(N):
    x,y = [],[]
    for i in range(1,N):
        xx = float(i)/float(N-1)
        x.append(xx)
        y.append(f(xx)+ np.random.normal(scale=0.2))
    return x,y

class Likelehood_fitting():
    def __init__(self,degree):
        self.degree = degree
        self.beta_=0.
        self.sigma = 0.
        self.w_ = None
    
    def func(self,x):
        r=0
        for i in range(self.degree + 1):
            r += x**i * self.w_[i]
        return r
    
    def fit(self,x,y):
        x,y = np.array(x),np.array(y)
        n = x.shape[0]
        phi = np.zeros([n,self.degree+1])
        for i in range(0,self.degree+1):
            phi[:,i] =  x**i
        
        A = np.dot(phi.T,phi)
        b = np.dot(phi.T,y)
        self.w_ = linalg.solve(A,b)
        #self.w_ = np.dot(np.linalg.inv(A),b)
        E = self.E(x,y)
        self.beta_ = n/E
        self.sigma = np.sqrt(1/self.beta_)
        
    def E(self,x,y):
        E = 0.
        for x_,t_ in zip(x,y):
            E += (self.func(x_) - t_)**2
        return E
    def log_likelibood(self,x,y):
        n = len(x)
        E = self.E(x,y)
        beta = n/E
        Log_p = 0.5*(-beta*E + n*np.log(beta/(2*np.pi)))
        return Log_p

    
trainx,trainy = dataset(10)
testx,testy = dataset(10)

fig = plt.figure(figsize=(15,9))

x = np.arange(0,1,0.01)
for i,d in enumerate([1,2,3,9]):
    model = Likelehood_fitting(d)
    model.fit(trainx,trainy)
    
    ax = fig.add_subplot(2,2,i+1)
    ax.set_xlim([-0.1,1.1])
    ax.set_ylim([-1.5,1.5])
    ax.scatter(trainx,trainy,c = "k",s=50)
    ax.plot(x,f(x),label = "True",c="b")
    ax.plot(x,model.func(x),label = f"pred, beta{round(model.sigma,1)}",c = "r")
    ax.fill_between(x, model.func(x)+model.sigma, model.func(x)-model.sigma,
                     color='c', alpha=0.5, label="predict_std")
   
    ax.legend(loc=1)
    
fig.show()
fig = plt.figure()
trlp,telp=[],[]
for i in range(0,10):
    model = Likelehood_fitting(i)
    model.fit(trainx,trainy)
    trlp.append(model.log_likelibood(trainx,trainy))
    telp.append(model.log_likelibood(testx,testy))
plt.plot(np.arange(0,10),trlp,label="train_likelhood")
plt.plot(np.arange(0,10),telp,label="test_likelhood")
plt.show()
```

### 変分線形回帰

```python
import matplotlib.pyplot as plt
import numpy as np
class Variational_Linear_Regression:
    def __init__(self,max_iter,degree):
        self.iter = max_iter
        self.degree = degree
        
    def Poly(self,X):
        return np.array([[x ** i for i in range(self.degree+1)] for x in X])
    
    def Gaussian_basis(self,X):
        bases = np.linspace(-1, 1, 5)
        return np.array([np.exp(- (X - b) ** 2. * 10.) for b in bases]).T

    def Repear(self,X):
#        phi = self.Gaussian_basis(X)
        phi = self.Poly(X)
        self.N,self.M = phi.shape
        self.m_N = np.zeros([self.M,1])
        self.S_N = np.eye(self.M)
        
        self.alpha_N,self.beta_N =.1,.1
        self.c_M,self.d_M =.1,.1
        self.E_alpha,self.E_beta = 0.001,100
        self.a0,self.b0 = 10.,10.
        self.c0,self.d0 = 1.,1.
        return phi
    
    def fit(self,X,t):
        Phi = self.Repear(X)
        
        for i in range(self.iter):
            params = np.hstack([self.E_alpha, self.E_beta])
            print(f"{i} : beta{self.E_beta}, alpha{self.E_alpha}")
            self.q_w(Phi,t)
            self.q_alpha()
            self.q_beta(Phi,t)
            
            if np.allclose(params,np.hstack([self.E_alpha, self.E_beta])):
                break
                
    def q_w(self,phi,t):
        S_inv = self.E_alpha*np.eye(self.M) + self.E_beta*(phi.T.dot(phi))
        self.S_N = np.linalg.inv(S_inv)
        self.m_N = (self.E_beta * self.S_N.dot(phi.T.dot(t))).reshape(-1,1)
    
        
    def q_alpha(self):
        E_ww = (self.m_N.T.dot(self.m_N) + np.trace(self.S_N)).flatten()
        self.alpha_N = self.a0 + self.M/2.
        self.beta_N = self.b0 + 0.5 * E_ww
        self.E_alpha = self.alpha_N/self.beta_N
    
    def q_beta(self,phi,t):

        self.c_M = self.c0 + 0.5*self.N
        a = (t.reshape(-1,1) - phi.dot(self.m_N))
        self.d_M = (self.d0 + 0.5*a.T.dot(a) \
                + np.trace(phi.T.dot(phi).dot(self.S_N))).flatten()
        self.E_beta = self.c_M/self.d_M
    
    def Predictive_distribution(self,X,t):
        #phi = self.Gaussian_basis(X)
        phi = self.Poly(X)
        
        mu = phi.dot(self.m_N)
        sigma = np.sqrt(1/self.E_beta + phi.dot(self.S_N.dot(phi.T)))
        
        return mu.flatten(),np.diag(sigma)



       
```

### 変分ロジスティック回帰

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import itertools
from functools import reduce


class PolynomialFeatures:

    def __init__(self, degree=2):
        self.degree = degree

    def transform(self, x):
        x_t = x.T
        features = [np.ones(len(x))]
        for degree in range(1, self.degree + 1):
            for items in itertools.combinations_with_replacement(x_t, degree):
                features.append(reduce(lambda x, y: x * y, items))
        return np.array(features).T


class Variational_Logistic_Regression():
    
    def __init__(self,max_iter):
        self.max_iter=max_iter
        
    def Repear(self,phi):
        self.N,self.M = phi.shape
        self.mu_N = np.zeros([self.M,1])
        self.Σ_N = np.eye(self.M)
        self.a_N,self.b_N =.1,.1
        self.E_alpha= 1.
        self.a0,self.b0 = 0,0
        self.xi=np.ones([self.M,self.M])
    
    def fit(self,Phi,t):
        self.Repear(Phi)
        for i in range(self.max_iter):
            params = np.hstack([self.E_alpha, self.xi.flatten()])
            self.q_alpha()
            self.q_w(Phi,t)
            self.opt_Xi(Phi)
         
            if np.allclose(params,np.hstack([self.E_alpha, self.xi.flatten()])):
                break
                
    def q_w(self,phi,t):
        self.mu_N = self.Σ_N.dot(phi.T.dot(t-1/2))
        self.Σ_N = np.linalg.inv(self.E_alpha*np.eye(self.M) + 2*self.lambda_(self.xi).dot(phi.T.dot(phi)))
        
    def q_alpha(self):
        self.a_N = self.a0 + self.M/2
        self.b_N = self.b0+0.5*np.trace(self.Σ_N + self.mu_N.T.dot(self.mu_N))
        self.E_alpha = self.a_N/self.b_N
        
    def opt_Xi(self,phi):
        xi = (phi.dot(self.Σ_N + self.mu_N.dot(self.mu_N.T))).T.dot(phi)
        self.xi = xi.dot(np.linalg.inv(xi))
        
    def lambda_(self,xi):
        return 0.5*xi.dot(self.sigmoid(xi) - 1/2)
    
    def sigmoid(self,x):
        return 1/(1 + np.exp(-x))
    
    def predict_dist(self,phi):
        mu = phi.dot(self.mu_N).flatten()
        sigma= np.diag(np.sqrt(phi.dot(self.Σ_N).dot(phi.T)))
        
        
        return self.sigmoid(mu / np.sqrt(1 + np.pi * sigma / 8))
    def predict_proba(self, X):
        return self.sigmoid(X.dot(self.mu_N))
    
    
from sklearn.model_selection import train_test_split
data = load_iris()
X = data.data[:100,[1,2]]
y = data.target[:100]

trainx,x_test,trainy,y_test = train_test_split(X,y,shuffle=True,random_state=100)
y=np.array(y).reshape(-1,1)

features = PolynomialFeatures(degree=3)
model = Variational_Logistic_Regression(10)
model.fit(feature.transform(X),y)


X_test, Y_test = np.meshgrid(np.linspace(0, 10, 100), np.linspace(0, 10, 100))
x_test = np.array([X_test, Y_test]).transpose(1, 2, 0).reshape(-1, 2)
probs = model.predict_dist(features.transform(x_test))
Probs = probs.reshape(100, 100)
dists = model.predict_dist(features.transform(x_test))
Dists = dists.reshape(100, 100)
levels = np.linspace(0, 1, 5)
cp = plt.contour(X_test, Y_test, Probs, levels, colors='k', linestyles='dashed')

colors = ['b', 'r']
plt.scatter(X[:, 0], X[:, 1], c=[colors[int(label)] for label in y])
plt.clabel(cp, inline=True, fontsize=10)
plt.contourf(X_test, Y_test, Dists, levels, alpha=0.5)
plt.colorbar()
plt.xlim(0, 10)
plt.ylim(0, 6)
plt.show()

np.sum(np.where(model.predict_proba(feature.transform(X))>=0.5,1,0) == y)/len(y)
```

```python
from scipy.special import digamma, gamma
def p(x):
    return np.sin(x*np.pi)**4
def sampling():
    k=1
    while 1:
        z=np.random.uniform(-1,1)
        u = np.random.uniform(0,k*0.5)
        if p(z) > u:
            return z

x=np.linspace(-1,1,200)
N = 1000000
p_chil = [sampling() for i in range(N)]
plt.plot(x,p(x))
plt.hist(p_chil,bins=50,normed=True)
plt.show()
```

#### 重点サンプリング and SIR

```python
from scipy.stats import *

# ガンマ分布のパラメータ
gamma_a = 5.0
gamma_b = 2

# 一様分布のパラメータ
uniform_loc = 0
uniform_scale = 15

# サンプルサイズ
N = 100000

def p(x):
    return gamma.pdf(x, a=gamma_a, scale=gamma_b)
    
def q(x):
    return uniform.pdf(x, loc=0, scale=15)
    
def sampling_q():
    return np.random.uniform(0,15,N)

def sampling_weighted(xs, weights):
    return np.random.choice(xs, size=N, p=weights)


# 分布q(x)からサンプリング
xs = sampling_q()

# 重みを計算
rates = np.array([p(x) / q(x) for x in xs])
rate_sum = rates.sum()
w = rates / rate_sum

# 離散確率分布から重み付けされた確率に従ってサンプリング
resample = sampling_weighted(xs, w)

# サンプリング結果のプロット
plt.hist(resample, bins=50, range=(uniform_loc, uniform_scale),
        color='C9',label='SIR', normed=1)

# 確率分布p(x)の描画
z = np.linspace(uniform_loc, uniform_scale, N)
plt.plot(z, p(z), color='blue')

plt.title('N = {}'.format(N))
plt.legend(loc='best', prop={'size': 9})
```

#### Markov chain hastings

```python
def q(x):
    z=np.random.normal(scale=0.5)
    return z+x

def p(x):
    if x < -1 or 1 < x:
        return 0
    return np.cos(0.5*np.pi*x)**2


def MCMC(x):
    while 1:
        z=q(x)
        if p(z)/p(x) >= np.random.uniform(0,1):
            return z
        
sample=[]
x=0.5
for i in range(100000):
    x=MCMC(x)
    sample.append(x)

x = np.linspace(-1, 1, 500)
y = np.vectorize(p)(x)
plt.hist(sample[1000:], bins=50, normed=1, alpha=0.5)
plt.plot(x, y)
plt.show()
```
